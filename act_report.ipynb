```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')
```

```python
df_master = pd.read_csv('twitter_archive_master.csv')
```

```python
#created a scatter plot of the favorite and retweet counts
df_master.plot(x = 'favorite_count', y = 'retweet_count', kind = 'scatter', alpha = 0.25)
plt.xlabel('Favorite Count')
plt.ylabel('Retweet COunt')
plt.title('Correlation between Retweet Count and Favorite Count')
```

```python
df_master.groupby('day_of_week')['tweet_id'].count().sort_values(ascending=False).plot(kind = 'barh')
plt.xlabel('Day of Week')
plt.ylabel('Number of Tweets')
plt.title('Number of Tweets by Day of week')
```
